```python
# -*- coding: utf-8 -*-
import os
import requests
import uuid
import json
import random

subscriptionKey = '8fca75eea9a942fbab9917743d164cc2'

base_url = 'https://api.cognitive.microsofttranslator.com/translate?api-version=3.0'
params = '&from=en&to=ko'
constructed_url = base_url + params

headers = {
    'Ocp-Apim-Subscription-Key': subscriptionKey,
    'Content-type': 'application/json',
    'X-ClientTraceId': str(uuid.uuid4())
}
```

```python
import codecs
data_list = json.load(codecs.open('new_processed_data.json', 'r', 'utf-8-sig'))
```

```python
data_list
set_data = set(data_list)
```

```python
for player_data in data_list:
    body = [{
                'text': player_data['name']
            }]
    request = requests.post(constructed_url, headers=headers, json=body)
    response = request.json()
    kor_name = response[0]['translations'][0]['text']
    korean = '' 
    for word in kor_name:
        if word != ' ':
            korean += word + ' '
    
    player_data['korean'] = korean.rstrip()
    print('{} --> {}'.format(player_data['name'], korean))
```

```python
with open('new_processed_data.json', 'w', encoding='UTF-8-sig') as json_file:  
    json.dump(data_list, json_file, indent=2, ensure_ascii=False)
```
